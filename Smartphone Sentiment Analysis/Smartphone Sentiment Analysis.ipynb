<b> Problem Explained </b>

Helio is working with a government health agency to create a suite of smartphone medical apps for use by aid workers in developing countries. This suite of apps will enable the aid workers to manage local health conditions by facilitating communication with medical professionals located elsewhere (one of the apps, for example, enables specialists in communicable diseases to diagnose conditions by examining images and other patient data uploaded by local aid workers).

Our job is to help Helio to understand what's the sentiment analysis towards the smartphones. Right now, they have a list of five potential devices to develop the app for, however they want to reduce that list to just one device. To do this, we will analyse the web and see what people are saying about these phones and for this, we need to go big with the data analysed.
To conduct this we are going to hire the cloud computing services from Amazon, the Elastic Compute Cloud and we will be using the data that is sourced by <a href="https://commoncrawl.org/"> Common Crawl. </a>

We will be using the Elastic Compute Cloud (EC2) to run applications programs in the Amazon computing environment, the Amazon Elastic MapReduce (EMR) that allow us to process a vast amount of data and we will store the data on the Amazon Simple Storage Services (S3).

To measure the sentiment analysis we will be looking and counting relevant words that help us to build a sentiment towards each phone, positive or negative references towards each phone or its OS.

There are also three files in python that are used for three different tasks:

<p><i>Mapper.py:</i> This mapper program is used to examine and count data from portions of the Common Crawl data.</p>
<p><i>Reducer.py:</i> This reducer program is used to accumulate the analysis from the individual mapper jobs.</p>
<p><i>Concatenatepv3.py:</i> This Python program is used to aggregate the results of your streaming jobs.</p>
